```python
import zipfile
from tqdm import tqdm
import os

def unzip_with_progress(zip_path, extract_to):
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        members = zip_ref.infolist()
        for member in tqdm(members, desc="Extracting", unit="file"):
            zip_ref.extract(member, extract_to)
```

```python
from paddleocr import PaddleOCR, draw_ocr
ocr = PaddleOCR(use_angle_cls=True, lang='japan', ocr_version = "PP-OCRv4") 
```

```python
from PIL import Image
import numpy as np

for i in range(10):
    img_path = f'datasets/validation/val-15k-1/i_s/0000{i}.png'
    img = Image.open(img_path)
    resized_img = img.resize((img.width * 4, img.height * 4))  # 2x scaling
    result = ocr.ocr(np.array(resized_img))
    # result = ocr.ocr(img_path, cls=True)
    
    if result == [[]]:
        print(f"\n\nFailed for 0000{i}.png")
        continue
    else:
        print(f"\n\nSucceeded for 0000{i}.png")
        print(result)
    for idx in range(len(result)):
        res = result[idx]
        for line in res:
            print(line)
```
